```python
import os
total_csv = os.listdir('C:\\Users\\rn0610\\Desktop\\P_folder\\P_W210Capstone\\finalcsv')
zipcode=[]
for i in total_csv:
    if i.endswith('csv'):
        zipcode.append(i.replace('.csv',''))
zipcode
#len(zipcode)
```

```python
import pandas as pd
columns=['imgname','lat','long','angle','class','confidence','xcenter','ycenter','width','hight','zipcode','zipcode2','detect_loc','new_lat','new_long','newzipcode']
df_empty=pd.DataFrame(columns=columns)
for i in range(len(zipcode)):
    df  = pd.read_csv('C:\\Users\\rn0610\\Desktop\\P_folder\\P_W210Capstone\\finalcsv\\'+zipcode[i]+'.csv')
    df['newzipcode']=zipcode[i]
    df_empty = df_empty.append(df, ignore_index=True)
df_empty
```

```python
df1 = df_empty[df_empty['imgname'].notnull()]
#80202, 80203, '80264','80290'
df1
```

```python
df1.to_csv('inference_csv_df.csv')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

zip_count = df1['newzipcode'].value_counts().sort_index()
# city_count = city_count[:10,]
plt.figure(figsize=(10,5))
chart = sns.barplot(zip_count.index, zip_count.values, alpha=0.8)
plt.title('Count of zipcode')
plt.ylabel('Number of Occurrences', fontsize=12)
plt.xlabel('zipcode', fontsize=12)
chart.set_xticklabels(chart.get_xticklabels(), rotation=45)
plt.show()
#80205 80202 are downtown
```

# Confidence level distribution for different classes

```python
classes=['lamp','signh','fh','nonpark','stop']
for i in range(5):
    cls = df1[df1['class']==i]['confidence'].value_counts(bins=25).sort_index()
    plt.figure(figsize=(10,5))
    chart = sns.barplot(cls.index, cls.values, alpha=0.8)
    plt.title('distribution of confidence '+classes[i])
    plt.ylabel('Number of Occurrences', fontsize=12)
    plt.xlabel(classes[i]+' confidence', fontsize=12)
    chart.set_xticklabels(chart.get_xticklabels(), rotation=90)
plt.show()
```

# Classes occurance distribution for different zipcode

```python
classes=['lamp','signh','fh','nonpark','stop']
for i in range(5):
    cls = df1[df1['class']==i]['newzipcode'].value_counts().sort_index()
    plt.figure(figsize=(10,5))
    chart = sns.barplot(cls.index, cls.values, alpha=0.8)
    plt.title(classes[i]+' zipcode distribution')
    plt.ylabel('Number of Occurrences', fontsize=12)
    plt.xlabel(classes[i]+' zipcode distribution', fontsize=12)
    chart.set_xticklabels(chart.get_xticklabels(), rotation=90)
plt.show()
```

# Make list for signh to plot signh user friendly area

```python
cls_signh = df1[df1['class']==1]['newzipcode'].value_counts().sort_index()
normalized_df=(cls_signh-cls_signh.min())/(cls_signh.max()-cls_signh.min())
df_concat = pd.concat([cls_signh, normalized_df], axis=1)
df_concat.columns=['num_signh','normalized_num_signh']
df_concat.to_csv('signh_zipcode_3.csv')
```
