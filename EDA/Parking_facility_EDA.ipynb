```python
df = pd.read_csv('parking_facilities.csv')
df.head()
```

```python
import matplotlib.pyplot as plt


city_count = df[df['DAILY_RATE'] !=-9999.0]['DAILY_RATE'].value_counts(bins=10).sort_index()
# city_count = city_count[:10,]
plt.figure(figsize=(10,5))
sns.barplot(city_count.index, city_count.values, alpha=0.8)
plt.title('Count of Daily rate')
plt.ylabel('Number of Occurrences', fontsize=12)
plt.xlabel('daily_rate', fontsize=12)
plt.show()
```

```python
parktype = df['PARKING_TYPE'].value_counts()
plt.figure(figsize=(10,5))
sns.barplot(parktype.index, parktype.values, alpha=0.8)
plt.title('Count of parking type')
plt.ylabel('Number of Occurrences', fontsize=12)
plt.xlabel('parking type', fontsize=12)
plt.show()
```

```python
parkmonthrate = df[df['MONTHLY_RATE'] !=-9999.0]['MONTHLY_RATE'].value_counts(bins=10).sort_index()
plt.figure(figsize=(10,5))
chart = sns.barplot(parkmonthrate.index, parkmonthrate.values, alpha=0.8)
plt.title('Count of parking monthly rate')
plt.ylabel('Number of Occurrences', fontsize=12)
plt.xlabel('Parking Monthly Rate', fontsize=12)
chart.set_xticklabels(chart.get_xticklabels(), rotation=45)
plt.show()
```
